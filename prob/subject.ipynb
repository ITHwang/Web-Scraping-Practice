```python
# wikipedia.org 사이트에서 '빅데이터'를 검색하기

from selenium import webdriver
from bs4 import BeautifulSoup as BS
import time

driver = webdriver.Chrome('./chrome/chromedriver')
url = 'https://www.wikipedia.org/'
driver.get(url)
time.sleep(3)
driver.find_element_by_xpath('//*[@id="searchInput"]').send_keys('빅데이터')
time.sleep(3)
driver.find_element_by_xpath('//*[@id="search-form"]/fieldset/button').click()
time.sleep(3)
defi = []
defi.append(driver.find_element_by_xpath('//*[@id="정의"]').text)
time.sleep(3)
defi.append(driver.find_element_by_xpath('//*[@id="mw-content-text"]/div[1]/p[6]').text)

for ele in defi:
    print(ele)
```

```python
#opinet 일반 휘발유 가격을 엑셀 파일로 저장하기

from urllib.request import urlopen
from selenium import webdriver
from bs4 import BeautifulSoup as BS
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

driver = webdriver.Chrome('./chrome/chromedriver')
time.sleep(3)
driver.get('https://www.opinet.co.kr')
time.sleep(3)
driver.find_element_by_xpath('//*[@id="quick_ul"]/li[2]/a').click()
time.sleep(3)
normal_gasoline = {}

driver.find_element_by_xpath('//*[@id="SIDO_NM0"]').send_keys('대전')
time.sleep(3)
_sigungu = driver.find_element_by_xpath('//*[@id="SIGUNGU_NM0"]')
time.sleep(3)
sigungu_ = _sigungu.find_elements_by_tag_name('option')
time.sleep(3)
sigungu = []
sigungu = [option.get_attribute('value') for option in sigungu_]
time.sleep(3)
for ele2 in sigungu:
    if ele2 != '':
        driver.find_element_by_xpath('//*[@id="SIGUNGU_NM0"]').send_keys(ele2)
        time.sleep(3)
        driver.find_element_by_id('searRgSelect').click()
        time.sleep(3)
        html = driver.page_source
        soup = BS(html, 'html.parser')
        name_ = soup.find('tbody',{'id':'body1'}).find_all('a')
        name = []
        for i in name_:
            name.append(i.text.strip())
            print(i.text)
        price_ = soup.find('tbody',{'id':'body1'}).find_all('tr')
        price = []
        for i in price_:
            price.append(i.find('td',{'class':'price'}).text.strip())
            print(i.find('td',{'class':'price'}).text)
        for i in range(len(name)):
            normal_gasoline[name[i]] = price[i]
        time.sleep(3)

columns = ['이름', '가격']
values = []
index = []
count = 1
for i in range(len(name)):
    index.append(count)
    value = []
    value.append(name[i])
    value.append(price[i])
    values.append(value)
    count += 1
    
df = pd.DataFrame(values, index = index, columns = columns)
df.to_excel('주유소.xlsx')
```
